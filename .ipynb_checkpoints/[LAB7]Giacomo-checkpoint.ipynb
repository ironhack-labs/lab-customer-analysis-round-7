```python
#data manipulation & operations
import pandas as pd
import numpy as np

import warnings

#plot
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

#statistical operation 
from sklearn.preprocessing import MinMaxScaler

import statsmodels.api as sm
from statsmodels.formula.api import ols
from sklearn import linear_model
#error metrics
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
analysis = pd.read_csv("/Users/giacomorossini/Desktop/IRONHACK/Laboratory/LAB5/lab-customer-analysis-round-5/files_for_lab/csv_files/marketing_customer_analysis.csv")
```

```python
#HEADINGS function
def transformation (df):
    cols=[]
    for i in range(len(df.columns)):
        cols.append(df.columns[i].lower().replace(' ','_'))
    df.columns=cols
    
transformation (analysis)
```

```python
#calling out numericals + categorical
numerical = analysis.select_dtypes(include = np.number)
categorical = analysis.select_dtypes(exclude = np.number)
numerical
```

```python
numerical.corr()['total_claim_amount'].sort_values()
```

```python
#check multicollinearity w/ Variance Inflaction Factor VIF

from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):
    vif = pd.DataFrame()
    vif['variables']= X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i)for i in range (X.shape[1])]
    display(vif.sort_values(by='VIF'))
```

```python
multic = numerical.drop(['total_claim_amount'], axis=1)
calc_vif(multic)
```

```python
#dropping monthly_premium_auto check VIF again
multic = numerical.drop(['monthly_premium_auto'], axis=1)
calc_vif(multic)
```

```python
sns.distplot(numerical['total_claim_amount'])
plt.xlabel('total_claim_amount')
plt.ylabel('Frequency')
plt.title('Total_claim_amount distribution')
plt.show()
```

```python
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(numerical)
normalized_data = pd.DataFrame(normalized_data, columns=numerical.columns)
numeric = normalized_data
```

```python
#encoded categorical

categ_encoded = pd.get_dummies(categorical.drop(['customer','effective_to_date','vehicle_class'], axis=1))
df_data = pd.concat([numeric,categ_encoded],axis=1)
```

```python
#dividing my dataframe --> train-test split: performing model on numerical dataset & whole data (future analysis)
from sklearn.model_selection import train_test_split

Y = df_data['total_claim_amount']
X = df_data.drop(['total_claim_amount'], axis = 1)


X_train, X_test, Y_train, Y_test = train_test_split(X , Y , test_size=0.2, random_state=42)
```

```python
X = sm.add_constant(X)
model = sm.OLS(Y,X).fit()
model.summary()
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
predictions = model.predict(X) 
print("R2 value is = ", round(r2_score(Y, predictions),2))

mse = mean_squared_error(Y, predictions)
mae = mean_absolute_error(Y, predictions)

print("The mean square error is =", mse)
print("The root mse model is =", np.sqrt(mse))
print("The mean absolute error is =", mae)
```

```python
#calculating the pvalues and drop >0.5 not relevant 

pvalues = pd.DataFrame(model.pvalues).rename(columns={0:'pvalue'})
pvalues['dipend_variable']=pvalues.index
pvalues=pvalues.reset_index().drop(['index'],axis=1)
pvalues.columns = ['pvalue', 'dipend_variable']
a = pvalues.sort_values(by='pvalue')
a
```

```python
condition = a['pvalue'] < 0.05
a_list = a[condition]['dipend_variable'].to_list()
a_list
```
