```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from math import ceil
import statsmodels.api as sm
from statsmodels.formula.api import ols
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
data= pd.read_csv(r"C:\Users\Quaresma\Documents\IRONHACK\Labs\7 lab-customer-analysis-round-7\files_for_lab\csv_files\marketing_customer_analysis.csv")
```

```python
col=[data.columns[i].lower().replace(" ","_") for i in range(len(data.columns))]
data.columns=col
data
```

```python
# try 
def cat_num(data):
    categoricals= data.select_dtypes(include=np.object )
    categoricals= categoricals.drop(columns=['customer','effective_to_date'])
    
    numerical= data.select_dtypes(include=np.number )
    
    return categoricals, numerical

categoricals, numerical = cat_num(data)
```

### NORMALIZATION

#### One Hot/Label Encoding (categorical)

```python
OHE=pd.get_dummies(categoricals, drop_first = True)
OHE
```

#### Normalize (numerical)

```python
from sklearn.preprocessing import StandardScaler

scaler1 = StandardScaler() #This is called initialization
stand_data = scaler1.fit_transform(numerical) # Fit into data
stand_data = pd.DataFrame(stand_data, columns= numerical.columns)
stand_data
```

```python
#---------------------------------------------------------------------------------------------------\\-------------------------------------------------------------------------------------------------------------------------------------
```

### Model with the features that have high correlation 

```python
X = data[['income','monthly_premium_auto']]
y = data.total_claim_amount

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python


lm= linear_model.LinearRegression()
model = lm.fit(X_train,y_train)
model.score(X_train,y_train) #This is the R2


predictions = lm.predict(X_test) 
print("R2 values is = ", round(r2_score(y_test, predictions),2))


mse= mean_squared_error(y_test,predictions)
mae= mean_absolute_error(y_test,predictions)
print("The mean squared error is = ", mse )
print("The mean absolute error is = ", mae )
print("The root mean squared error is = ", np.sqrt(mse) )
```
